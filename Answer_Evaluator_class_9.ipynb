```python
# for class 10 and class 5 (BPNB), class 5 (BKBV), class 6 (BKBV), class 7 (BKBV), class 8 (BKBV), class 10 (BKBV) "ans" is small letters
def just_eval(metrics_list, ans):
  print(ans)
  if (ans == 'a'):
    for b in range(num_bright):
      if (metrics_list[b] != 'n'):
        metrics_list[b] = float(metrics_list[b]) + 2
      else:
        metrics_list[b] = 0
    for d in range(num_bright, num_bright + num_dark):
      if (metrics_list[d] != 'n'):
        metrics_list[d] = float(metrics_list[d]) - 2
      else:
        metrics_list[d] = 0

  elif (ans == 'b'):
    for b in range(num_bright):
      if (metrics_list[b] == 'n'):
        metrics_list[b] = 0
    for d in range(num_bright, num_bright + num_dark):
      if (metrics_list[d] == 'n'):
        metrics_list[d] = 0

  elif (ans == 'c'):
    for b in range(num_bright):
      if (metrics_list[b] != 'n'):
        metrics_list[b] = float(metrics_list[b]) - 2
        if metrics_list[b] < 0:
          metrics_list[b] = 0
      elif (metrics_list[b] == 'n'):
        metrics_list[b] = 0
    for d in range(num_bright, num_bright + num_dark):
      if (metrics_list[d] != 'n'):
        metrics_list[d] = float(metrics_list[d]) + 2
        if metrics_list[d] > 10:
          metrics_list[d] = 10
      elif (metrics_list[d] == 'n'):
        metrics_list[d] = 0

  elif (ans == 'd'):
    for b in range(num_bright):
        metrics_list[b] = 0
    for d in range(num_bright, num_bright + num_dark):
      if (metrics_list[d] != 'n'):
        metrics_list[d] = 10
      else:
        metrics_list[d] = 0

  elif (ans == 'e'):
    for b in range(num_bright):
      if (metrics_list[b] != 'n'):
        metrics_list[b] = float(metrics_list[b]) - 1.5
        if metrics_list[b] < 0:
          metrics_list[b] = 0
      elif (metrics_list[b] == 'n'):
        metrics_list[b] = 0
    for d in range(num_bright, num_bright + num_dark):
      if (metrics_list[d] != 'n'):
        metrics_list[d] = float(metrics_list[d]) + 2
        if (metrics_list[d] > 10):
          metrics_list[d] = 10
      elif (metrics_list[d] == 'n'):
        metrics_list[d] = 0

  return metrics_list
```

```python
def find_indices(list_to_check, item_to_find):
  indices = []
  for idx, value in enumerate(list_to_check):
    if value == item_to_find:
      indices.append(idx)
  return indices
```

```python
for s in range(df.shape[0]):

  print(s)

  divisors = [num_ques] * (num_bright + num_dark)

  for q in range(1,num_ques + 1):

    print("ques_no =", q)

    index = 2*q - 1
    print("option_selected = ", df.iloc[s,index])

    #MCQ
    if (df.iloc[s,index] != "None"):
      string_ = df_.iloc[q-1, int(df.iloc[s,index])]
      score_list = [str(item) for item in string_.split(',')]

      for index_ in find_indices(score_list, 'n'):
        divisors[index_] = divisors[index_] - 1

      print(score_list)
      if q not in ques_no_just:
        print("just_ans = ", df.iloc[s,index+1])
        score_list = just_eval(score_list, df.iloc[s, index + 1])

      if q in ques_no_just:
        for j in range(len(score_list)):
          if (score_list[j] == 'n'):
            score_list[j] = 0

      #score_list  = [float(x) for x in score_list]
      for j in range(len(score_list)):
          if score_list[j] == 'n':
             score_list[j] = 0.0  # Replace 'n' with 0.0 or any other suitable numeric value
          else:
             score_list[j] = float(score_list[j])
      print(score_list)
      #for p in range(len(score_list)):
        #df.iloc[s, 2*num_ques + 1 + p] = df.iloc[s, 2*num_ques + 1 + p] + score_list[p]
      for p in range(len(score_list)):
          if df.iloc[s, 2*num_ques + 1 + p] != 'None':
              df.iloc[s, 2*num_ques + 1 + p] = float(df.iloc[s, 2*num_ques + 1 + p]) + score_list[p]
          else:
              # Handle the case where the value is 'None', e.g., by setting it to 0 or another appropriate value.
              df.iloc[s, 2*num_ques + 1 + p] = 0.0  # Or any other value you prefer

    #unattempted
    elif (df.iloc[s,index] == "None"):
      string__ = df_.iloc[q-1, 5]
      score_list_ = [float(item) for item in string__.split(',') if item.isdigit()]
      print(score_list_)
      for p in range(len(score_list_)):
        df.iloc[s, 2*num_ques + 1 + p] = df.iloc[s, 2*num_ques + 1 + p] + score_list_[p]

    print('----------------------')

  for tn in range(len(score_list)):
    df.iloc[s, 2*num_ques + 1 + tn] = df.iloc[s, 2*num_ques + 1 + tn] / divisors[tn]
```
